```python
import pandas as pd

# Add a column to each dataframe to mark the meal type
df_breakfast['Meal_Type'] = 'Breakfast'
df_lunch['Meal_Type'] = 'Lunch'

# Merge (stack) them together
df_combined = pd.concat([df_breakfast, df_lunch], ignore_index=True)

# Check the result
print(df_combined['Meal_Type'].value_counts())
print(df_combined.shape)
```

```python
df_combined.info()
```

```python
df_combined.head(20)
```

```python
cost_cols = [
    'Served_Non-Reimbursable', 'Discarded_Total', 'Discarded_Cost',
    'Subtotal_Cost', 'Left_Over_Percent_of_Offered',
    'Left_Over_Cost', 'Production_Cost_Total'
]

for col in cost_cols:
    df_combined[col] = (df_combined[col].astype(str)
                        .str.replace('[\$,%,]', '', regex=True)
                        .replace(['', 'NA', 'nan', 'None'], pd.NA))
    df_combined[col] = pd.to_numeric(df_combined[col], errors='coerce')
df_combined.describe()
```

```python
import pandas as pd
import numpy as np
df_combined['Date'] = pd.to_datetime(df_combined['Date'])
num_cols = [
    'Served_Total','Offered_Total','Planned_Total',
    'Discarded_Cost','Production_Cost_Total','Left_Over_Total'
]

for col in num_cols + [
    'Served_Non-Reimbursable','Discarded_Total',
    'Subtotal_Cost','Left_Over_Percent_of_Offered','Left_Over_Cost'
]:
    df_combined[col] = (df_combined[col].astype(str)
                        .str.replace('[\$,%,]', '', regex=True)
                        .replace(['', 'NA', 'nan', 'None'], pd.NA))
    df_combined[col] = pd.to_numeric(df_combined[col], errors='coerce')
df_combined['offered_safe'] = np.where(df_combined['Offered_Total']>0,
                                       df_combined['Offered_Total'], np.nan)
```

```python
df_combined['waste_rate'] = df_combined['Discarded_Cost'] / df_combined['offered_safe']
df_combined['cost_per_served'] = df_combined['Production_Cost_Total'] / df_combined['Served_Total']
df_combined['served_rate'] = df_combined['Served_Total'] / df_combined['offered_safe']
```

```python
school_summary = (df_combined
    .groupby('School_Name')
    .agg({
        'Served_Total':'sum',
        'Discarded_Cost':'sum',
        'Production_Cost_Total':'sum',
        'offered_safe':'sum'
    }))

school_summary['waste_rate'] = school_summary['Discarded_Cost']/school_summary['offered_safe']
school_summary['cost_per_served'] = school_summary['Production_Cost_Total']/school_summary['Served_Total']

school_summary.sort_values('waste_rate').head(10)  # lowest waste schools
school_summary.sort_values('cost_per_served').head(10)  # most cost-efficient schools
```

```python
item_summary = (df_combined
    .groupby('Identifier')
    .agg({
        'Served_Total':'sum',
        'Discarded_Cost':'sum',
        'Production_Cost_Total':'sum',
        'offered_safe':'sum'
    }))

item_summary['waste_rate'] = item_summary['Discarded_Cost']/item_summary['offered_safe']
item_summary['cost_per_served'] = item_summary['Production_Cost_Total']/item_summary['Served_Total']

item_summary.sort_values('waste_rate').head(10)      # low waste items
item_summary.sort_values('cost_per_served').head(10) # cost-efficient items
item_summary.sort_values('Served_Total', ascending=False).head(10) # most popular items
```

```python
# Add time features
df_combined['dow'] = df_combined['Date'].dt.day_name()
df_combined['month'] = df_combined['Date'].dt.month

# By Meal Type
df_combined.groupby('Meal_Type')[['waste_rate','cost_per_served','served_rate']].mean()

# By Day of Week
df_combined.groupby('dow')[['waste_rate','cost_per_served','served_rate']].mean().sort_values('waste_rate')

# By Month
df_combined.groupby('month')[['waste_rate','cost_per_served','served_rate']].mean().sort_index()
```

```python
df_combined.groupby('FCPS Region')[['waste_rate','cost_per_served']].mean()
df_combined.groupby('Level')[['waste_rate','cost_per_served']].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(
    item_summary['cost_per_served'],
    item_summary['waste_rate'],
    alpha=0.5
)
plt.xlabel('Cost per Served Meal')
plt.ylabel('Waste Rate (Discarded Cost / Offered)')
plt.title('Menu Items: Cost vs Waste')
plt.axhline(item_summary['waste_rate'].mean(), color='r', linestyle='--', label='Avg Waste')
plt.axvline(item_summary['cost_per_served'].mean(), color='g', linestyle='--', label='Avg Cost')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    item_summary['Served_Total'],
    item_summary['waste_rate'],
    s=item_summary['Served_Total']/100,  # scale size
    alpha=0.4
)
plt.xlabel('Total Served')
plt.ylabel('Waste Rate')
plt.title('Menu Items: Popularity vs Waste')
plt.show()
```
